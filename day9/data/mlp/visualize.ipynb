```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot some examples of the latent 2D space.

```python
def plot_latent_space(x):
    b = x.reshape(-1, 28, 28)
    canvas = np.empty((28 * 21, 28 * 21))

    plt.figure(figsize=(21,21))

    for j in range(21):
        for i in range(21):
            canvas[j * 28:(j + 1) * 28, i * 28:(i + 1) * 28] = b[i + 21 * j]

    plt.imshow(canvas)
    plt.show()
```

```python
plot_latent_space(np.loadtxt("latent_space_2D.txt"))
```

Plot color points representing all test set in encoded 2D space.

```python
betas = [0,1,4]

for beta in betas:
    filename = "beta_%d.log" % beta

    dta = np.loadtxt(filename)
    print(filename, dta.shape)
    plt.figure(figsize=(10,10))

    plt.title(r"$\beta$=%d" % beta)
    for label in range(10):
        indices = np.where(dta[:,0] == float(label))
        plt.scatter(dta[indices, 1], dta[indices, 2], label=label)
    plt.legend()
    plt.show()
```
